```python
import matplotlib.pyplot as plt
import numpy as np

# Simple arrays for the plot
x = np.array([0, 2, 4, 6, 8, 10])
y1 = np.array([0, 1, 0, -1, 0, 1])
y2 = np.array([1, 0, -1, 0, 1, 0])
error = [0.1, 0.2, 0.1, 0.2, 0.1, 0.2]  # Error values for error bars

# Line plots with labels
plt.plot(x, y1, label='sin(x)', c='blue', ls='-', lw=3, marker='o', ms=6, mec='blue',mfc='blue')
plt.plot(x, y2, label='cos(x)', c='red', ls='--', lw=3, marker='s', ms=6)

# Add a grid
plt.grid(color='gray', linestyle='--', linewidth=0.5, alpha=0.7)

# Add error bars
plt.errorbar(x, y1, yerr=error, fmt='o', color='blue', ecolor='black', elinewidth=1, capsize=3, label='sin(x) error')
plt.errorbar(x, y2, yerr=error, fmt='s', color='green', ecolor='green', elinewidth=1, capsize=3, label='cos(x) error')

# Add titles and labels
plt.title('Line Plot', fontsize=16, fontweight='bold')
plt.xlabel('X-Axis', fontsize=14, fontweight='bold')
plt.ylabel('Y-Axis', fontsize=14, fontweight='bold')

# Customize ticks
plt.xticks(range(0, 11, 2), fontsize=12, color='darkred')
plt.yticks([-1, -0.5, 0, 0.5, 1], fontsize=12, color='darkblue')

# Add annotations
plt.annotate('Peak', xy=(2, 1), xytext=(3, 1.2),
             arrowprops=dict(facecolor='black', arrowstyle='->'), fontsize=12)

# Add a legend
plt.legend(title='Functions', fontsize=12, title_fontsize=14, loc='upper right', frameon=True)

# Set plot limits
plt.xlim(0, 10)
plt.ylim(-1.5, 1.5)

# Add text inside the plot
plt.text(7, 1.2, 'Matplotlib Demo', fontsize=14, color='purple', style='italic')

# Display the plot
plt.show()
```

```python
# Data for the pie chart
data = np.array([30, 25, 50, 45, 10])  # Values for each slice
mylabels = np.array(['A', 'B', 'C', 'D', 'E'])
mycolors = np.array(['#1f77b4', '#ff7f0e', '#2ca02c', '#d62728', '#9467bd'])  # Dark colors
myexplode = np.array([0.1, 0.1, 0, 0, 0.2])  # Highlight slices A, B, and E

plt.pie(
    data,
    explode=myexplode,
    labels=mylabels,
    colors=mycolors,
    autopct='%1.0f%%',
    startangle=90,
    pctdistance=0.85,
    shadow=True  # Adds a shadow effect
)

# Add annotations
plt.annotate(
    'Largest Slice',
    xy=(0.8, 0.3),  
    xytext=(0.8, 0.9),
    arrowprops=dict(facecolor='black', arrowstyle='->'),
    fontsize=12, color='black'
)

# Add a legend outside the chart
plt.legend(title="Categories", loc="center left", fontsize=10)

# Add a title
plt.title('Pie Chart', fontsize=18, fontweight='bold', color='black')

# Display the plot
plt.show()
```

```python
# Data for the scatter plot
x = np.random.randint(high=25,low=1,size=50) # Random x values
y = np.random.randint(high=25,low=1,size=50)  # Random y values
sizes = np.random.randint(low = 10, high = 500, size = 50)  # Random sizes for bubbles
colors = np.random.randint(high=25,low=1,size=50)  # Random colors

# Create scatter plot
scatter=plt.scatter(
    x, y, 
    s=sizes, 
    c=colors, 
    cmap='viridis', 
    alpha=0.8, 
    edgecolor='black', 
    label='Data Points',
    marker='*'
)

# Adding color bar
cbar = plt.colorbar(scatter)
cbar.set_label('Color Intensity', fontsize=12)

# Adding grid
plt.grid(True)

# Adding titles and labels
plt.title('Scatter Plot', fontsize=18, fontweight='bold')
plt.xlabel('X Values', fontsize=12)
plt.ylabel('Y Values', fontsize=12)

# Display the plot
plt.show()
```

```python
# Data for the bar chart
categories = ['A', 'B', 'C', 'D', 'E']
values = [30, 25, 50, 45, 10]
colors = ['#1f77b4', '#ff7f0e', '#2ca02c', '#d62728', '#9467bd'] 

# Create subplots
plt.subplot(1,2,1)

# Vertical bar chart
plt.bar(categories, values, color=colors, edgecolor='black', linewidth=1.2)
plt.title('Vertical Bar Chart', fontsize=16, fontweight='bold')
plt.xlabel('Categories', fontsize=12)
plt.ylabel('Values', fontsize=12)
plt.grid(axis='y', linestyle='--', alpha=0.7)

plt.subplot(1,2,2)
# Horizontal bar chart
plt.barh(categories, values, color=colors, edgecolor='black', linewidth=1.2)
plt.title('Horizontal Bar Chart', fontsize=16, fontweight='bold')
plt.xlabel('Values', fontsize=12)
plt.ylabel('Categories', fontsize=12)
plt.grid(axis='x', linestyle='--', alpha=0.7)


# Add a super title
plt.suptitle('Comprehensive Bar Chart', fontsize=20, fontweight='bold')

# Adjust layout
plt.tight_layout()

# Display the plot
plt.show()
```

```python
# Data for the histogram
data = np.random.randint(low=10,high=90,size=100) 
print(data)

plt.hist(
    data, 
    bins=10, 
    color='darkblue', 
    edgecolor='black', 
    alpha=0.7, 
    label='Random Data', 
    histtype='bar',
    orientation = 'vertical',
    align = 'mid'
)

# Adding mean and median lines
mean = np.mean(data)
median = np.median(data)
plt.axvline(mean, color='red', linestyle='--', linewidth=2)
plt.axvline(median, color='white', linestyle='-.', linewidth=2)

# Adding grid
plt.grid(linestyle='--', alpha=0.7)

# Adding titles and labels
plt.title('Histogram', fontsize=18, fontweight='bold')
plt.xlabel('Data Values', fontsize=12)
plt.ylabel('Frequency', fontsize=12)

# Adjust layout
plt.tight_layout()

# Display the plot
plt.show()
```

```python
# Data for the area plot
x = np.linspace(0, 10, 100)
print(x)
y1 = np.sin(x)
y2 = np.cos(x)

plt.subplot(2,1,1)
# Plot the area chart
plt.fill_between(x, y1, color='blue', alpha=0.7, label='Dataset 1',hatch='//',edgecolor='black')
plt.fill_between(x, y2, color='purple', alpha=0.7, label='Dataset 2',edgecolor='black')

# Add grid
plt.grid(True, linestyle='--', linewidth=0.5, alpha=0.7)

plt.xlabel('X-axis', fontsize=12)
plt.ylabel('Y-axis', fontsize=12)

plt.subplot(2,1,2)
plt.fill_between(x, y1, y2, where=(y1 > y2), color='green', alpha=0.3, label='Y1 > Y2')
plt.fill_between(x, y1, y2, where=(y1 <= y2), color='red', alpha=0.3, label='Y2 >= Y1')

plt.grid(True, linestyle='--', linewidth=0.5, alpha=0.7)

plt.title('Area Plot', fontsize=18, fontweight='bold')
plt.xlabel('X-axis', fontsize=12)
plt.ylabel('Y-axis', fontsize=12)

plt.suptitle('Area Plot', fontsize=18, fontweight='bold')
plt.tight_layout()

# Display the plot
plt.show()
```

```python
# Data for heatmaps
data1 = np.random.rand(10, 10)  # Random data for heatmap 1
data2 = np.random.rand(10, 10)  # Random data for heatmap 2
data3 = np.random.rand(10, 10)  # Random data for heatmap 3
data4 = np.random.rand(10, 10)  # Random data for heatmap 4

# Heatmap 1
plt.subplot(2, 2, 1)
plt.imshow(data1, cmap='viridis')
plt.title("Heatmap 1: Viridis")
plt.colorbar(label='Scale')

# Heatmap 2
plt.subplot(2, 2, 2)
plt.imshow(data2, cmap='magma')
plt.title("Heatmap 2: Magma")
plt.colorbar(label='Intensity')

# Heatmap 3
plt.subplot(2, 2, 3)
plt.imshow(data3, cmap='coolwarm')
plt.title("Heatmap 3: Coolwarm")
plt.colorbar(label='Value')

# Heatmap 4
plt.subplot(2, 2, 4)
plt.imshow(data4, cmap='cividis')
plt.title("Heatmap 4: Cividis")
plt.colorbar(shrink=0.8, label='Value')

# Adjust layout
plt.tight_layout()

# Display the plot
plt.show()
```

```python

# Data
data1 = np.array([1, 2, 3, 4, 5])
data2 = np.array([2, 4, 3, 5, 6])
data3 = np.array([3, 5, 4, 2, 4])

# Creating a stack plot
plt.stackplot(
    data1,data2,data3,
    labels=['A','B','C'], 
    colors=['green', 'purple', 'hotpink'], 
    edgecolor='black',
    alpha=0.8
)

plt.grid(True)

# Adding labels and title
plt.xlabel('X axis', fontsize=12)
plt.ylabel('Y axis', fontsize=12)
plt.title('Stack Plot', fontsize=16)

# Displaying the stack plot
plt.tight_layout()
plt.show()
```

```python
import pandas as pd

# Generating time-series data
dates = pd.date_range(start="2023-01-01", periods=100, freq="D")
data1 = np.random.rand(100) * 10  # Random data for Line 1
data2 = np.random.rand(100) * 5   # Random data for Line 2

# Plot Line 1
plt.plot(dates, data1, label='Dataset 1', color='blue', linestyle='-', marker='o', markersize=4)

# Plot Line 2
plt.plot(dates, data2, label='Dataset 2', color='green', linestyle='--', marker='s', markersize=4)

# Highlight a specific range
plt.axvspan(dates[20], dates[40], color='gray', alpha=0.2, label='Highlighted Range')

# Adding grid lines
plt.grid(color='lightgray', linestyle='--', linewidth=0.5)

# Adding labels, title, and legend
plt.xlabel('Date', fontsize=12)
plt.ylabel('Value', fontsize=12)
plt.title('Time-Series Data Plot', fontsize=16, fontweight='bold')
plt.legend(fontsize=10, loc='upper right')

# Rotating x-axis labels for better visibility
plt.xticks(rotation=45)

# Adjust layout
plt.tight_layout()

# Display the plot
plt.show()
```

```python
from mpl_toolkits.mplot3d import Axes3D

x = np.linspace(-5, 5, 100)
y = np.linspace(-5, 5, 100)
x, y = np.meshgrid(x, y)
z = np.sin(np.sqrt(x**2 + y**2))

ax = plt.subplot(111, projection='3d') 

# 3D surface plot
surf = ax.plot_surface(x, y, z, cmap='viridis', edgecolor='k')

# Adding color bar
plt.colorbar(surf, shrink=0.5, aspect=10) 

# Adding title and labels
plt.title('3D Surface Plot', fontsize=16, fontweight='bold')
ax.set_xlabel('X-axis', fontsize=12)
ax.set_ylabel('Y-axis', fontsize=12)
ax.set_zlabel('Z-axis', fontsize=12)

plt.tight_layout()
plt.show()
```

```python

# Generating data
theta = np.linspace(0, 2 * np.pi, 100)
r = np.abs(np.sin(2 * theta) * np.cos(3 * theta))

# Creating polar plot
plt.subplot(111, polar=True)
plt.plot(theta, r, color='red', linewidth=2.5, label='Polar Curve')

# Adding title
plt.title('Polar Plot', fontsize=16, fontweight='bold', pad=30)

plt.tight_layout()
plt.show()
```
